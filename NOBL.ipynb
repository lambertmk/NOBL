```python
import numpy as np # analysis
import pandas as pd # for dataframe
import matplotlib.pyplot as plt # plotting
import seaborn as sns # fancy plotting
from scipy import stats # testing
```

```python
data = pd.read_csv('NOBL Questions from adapt survey.csv')
```

```python
data
```

```python
data['Which letter most closely represents where you are on this cloud journey?'] = data['Which letter most closely represents where you are on this cloud journey']
```

```python
# create new column for letter only

data.letter = data['Which letter most closely represents where you are on this cloud journey?'].astype(str).str[0]
data.letter.value_counts()
```

```python
sns.set_theme(style="ticks")
sns.color_palette("crest_r", as_cmap=True)

f, ax = plt.subplots(figsize=(7, 5))
sns.despine(f)

sns.histplot(
    data,
    x=data.letter)
```

```python
sns.set_theme(style="whitegrid")

f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.scatterplot(x=data.letter, y=data.Employees,
                data=data, ax=ax)

# no clear relationship here
```

```python
sns.set_theme(style="whitegrid")

f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.scatterplot(y=data.letter, x=data.Employees,
                data=data, ax=ax)

# same when the axes are switched
```

```python
sns.set_theme(style="ticks")

f, ax = plt.subplots(figsize=(7, 5))
sns.despine(f)

sns.histplot(
    data,
    x=data.letter, hue=data.Sector,
    multiple="stack",
    palette="crest_r")
```
